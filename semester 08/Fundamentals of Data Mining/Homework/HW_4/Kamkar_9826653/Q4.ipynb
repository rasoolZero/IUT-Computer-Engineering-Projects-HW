```python
#1
import pandas as pd
df = pd.read_csv('data.txt',header=None)
df.head()
```

```python
#2
df.describe()
```

```python
#3
for column in df.columns:
    print(f"number of zeros in column {column} is :{(df[column] == 0).sum()}")
```

#### working on columns 1 to 5 (zero indexed)

```python
#4 and 5
import numpy as np
for column in range(1,6):
    df[column] = df[column].replace(0,np.nan)
for column in range(1,6):
    print(f"number of zeros in column {column} is :{(df[column].isnull()).sum()}")
```

```python
#6
df.head(15)
```

```python
#7
df2 = df.iloc[:,1:6]
df2.dropna(inplace=True)

print(f"rows before dropping nan : {df.shape[0]}")
print(f"rows after dropping nan : {df2.shape[0]}")
```

```python
#8
df3 = df.iloc[:,1:6]
df3.fillna(df3.mean().mean(),inplace=True)
for column in range(1,6):
    print(f"number of zeros in column {column} is : {(df3[column].isnull()).sum()}")
df3.head(10)
```

```python
#9
from sklearn.impute import SimpleImputer
df4 = df.iloc[:,1:6]
imputer = SimpleImputer(strategy='mean')
df4 = pd.DataFrame(imputer.fit_transform(df4),columns=[1,2,3,4,5])
for column in range(1,6):
    print(f"number of zeros in column {column} is : {(df4[column].isnull()).sum()}")
df4.head(10)
```

```python
#10 - median
dfmedian = df.iloc[:,1:6]
imputer = SimpleImputer(strategy='median')
dfmedian = pd.DataFrame(imputer.fit_transform(dfmedian),columns=[1,2,3,4,5])
for column in range(1,6):
    print(f"number of zeros in column {column} is : {(dfmedian[column].isnull()).sum()}")
dfmedian.head(10)
```

```python
#10 - most_frequent
dffrequent = df.iloc[:,1:6]
imputer = SimpleImputer(strategy='most_frequent')
dffrequent = pd.DataFrame(imputer.fit_transform(dffrequent),columns=[1,2,3,4,5])
for column in range(1,6):
    print(f"number of zeros in column {column} is : {(dffrequent[column].isnull()).sum()}")
dffrequent.head(10)
```
