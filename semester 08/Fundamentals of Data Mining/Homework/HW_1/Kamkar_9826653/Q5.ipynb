```python
#1
import pandas as pd
dataframe = pd.read_csv("Smartphone.csv",delimiter=",")
dataframe
```

```python
#2
dataframe.info()
```

```python
#3
bluetooth_10GB = dataframe[(dataframe["blue"]==1) & (dataframe["int_memory"]>10)]
bluetooth_10GB
```

```python
#4-1
bestfrontcameras = dataframe[(dataframe.fc == dataframe["fc"].max())]
bestfrontcameras
```

```python
#4-2
worstfrontcameras = dataframe[(dataframe.fc == dataframe["fc"].min())]
worstfrontcameras
```

```python
#5
def clockrate(speed:int) -> str:
    if speed <=1:
        return "Low"
    if speed <=2:
        return "Mid"
    else:
        return "High"
    
dataframe['clock_rate'] = dataframe['clock_speed'].apply(lambda x: "Low" if x <= 1 else "Mid" if x <= 2 else "High")
dataframe
```

```python
#6-heatmap
import seaborn as sns, matplotlib.pyplot as plt
correlations = dataframe.corr(numeric_only=True)
fig, ax = plt.subplots(figsize=(10,10)) 
sns.heatmap(correlations,square=True,ax = ax)
```

## report

**based on the heatmap (except for price_range), we can see that these pairs of attributes are linearly related:**

- Primary camera quality (pc) - Front camera quality (fc)
- 3g - 4g
- screen height (sc_h) - screen width(sc_w)
- pixel resolution height (px_height) - pixel resolution width (px_width)
