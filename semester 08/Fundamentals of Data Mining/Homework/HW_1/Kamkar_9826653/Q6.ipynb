```python
import pandas as pd
```

```python
#1
df = pd.read_csv("Student.csv",delimiter=",")
df
```

```python
#2_histogram
df["math score"].hist(by = df["gender"],sharex = True,sharey = True)
#2_gender_describe
for gender in ["male","female"]:
    selected_df = df[["gender","math score"]]
    gender_df = selected_df[df["gender"]==gender]
    describe_df = gender_df.describe()
    print("\ngender : ",gender)
    print(describe_df.loc[["mean","min","max"]])
```

```python
df["race/ethnicity"].value_counts().plot.pie()
```
